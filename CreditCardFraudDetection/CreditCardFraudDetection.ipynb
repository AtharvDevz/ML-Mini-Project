```python
# Extracting the zip file
from zipfile import ZipFile
dataset = "/content/creditcardfraud.zip"

with ZipFile(dataset, 'r') as zip:
  zip.extractall()
  print("dataset is Extraxted")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(dataset)
```

```python
# Sepreting data for Analysis
legit = data[data.Class == 0]
print(legit.shape)
fraud = data[data.Class == 1]
print(fraud.shape)
```

UnderSampling Build Sample dataset Containg Similar Distributon of Normal transaction and Fraud Transaction..

```python
legit_sample = legit.sample(n=492)
```

```python
# Concate Two DataFrame
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
# Split Data Train and Test DAta
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state = 2)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# MOdel Training
model = LogisticRegression()
```

```python
# Training Model With the Data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
```

```python
X_train_prediction = model.predict(X_train)
accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
accuracy
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
test_accuracy
```

```python
import pickle
filename = "FraudDetectionModel.sav"
Model = pickle.dump(model , open(filename, 'wb'))
```
