```python
from sklearn.datasets import load_breast_cancer
breast_cancer_data = load_breast_cancer()
```

```python
breast_cancer_data
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(data=breast_cancer_data.data,
                  columns= breast_cancer_data.feature_names)
```

```python
df.info()
```

```python
df.shape
```

```python
df['label'] = breast_cancer_data.target
```

```python
df.shape
```

```python
df.describe()
```

```python
df['label'].value_counts()
```

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
my_pipeline = Pipeline([
    ('pca', PCA(n_components=2))
])
```

```python
x = my_pipeline.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=breast_cancer_data.target, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatterplot of PCA Data')
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, random_state=42)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
## Model Evaluation
from sklearn.metrics import accuracy_score
```

```python
X_train_pred = model.predict(X_train)
train_data_accuracy = accuracy_score(Y_train, X_train_pred)
print("Accuracy : ", train_data_accuracy)
```

```python
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_pred)
print("Acccuracy : ", test_data_accuracy)
```

```python
model2 = LogisticRegression()
model2.fit(x_train, y_train)
```

```python
x_train_pred = model2.predict(x_train)
```

```python
x_train_accuracy = accuracy_score(y_train, x_train_pred)
print("Accuracy : ", x_train_accuracy)
```

```python
x_test_pred = model2.predict(x_test)
x_test_accuracy = accuracy_score(y_test, x_test_pred)
print("Accuracy : ", x_test_accuracy)
```
